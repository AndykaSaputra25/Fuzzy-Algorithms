```python
import numpy as np
import skfuzzy as fuzzy
import matplotlib.pyplot as plt
%matplotlib inline
```

Jika Terdapat fungsi keanggotaan dari suatu himpunan A yang terlihat seperti gambar di bawah ini, maka tentukan nilai derajat keanggotaan dari 0,x = 60, x=70, x= 90, dan 10

```python
x = np.arange(0, 120)
himpunanA = fuzzy.trapmf(x, [60,100,120,120])

fig, ax = plt.subplots(nrows=1, figsize=(8,5))
ax.plot(x, himpunanA, 'b', linewidth=4)

plt.title('Himpunan A')
plt.xlabel('X')
plt.ylabel('Derajat Keanggotaan')

plt.show()
```

```python
data = [0, 60, 70, 90, 110, 10] # Soal

for i in range(len(data)):
    hasil = fuzzy.interp_membership(x, himpunanA, data[i]) 
    print('Derajat Keanggotaan dari ', data[i], ' = ', hasil)
```
